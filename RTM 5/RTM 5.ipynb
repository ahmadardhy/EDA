```python
p_x_c_2_mamalia = 0 * 0 * 0 * 0.958333 * 0.75 * 0.5 * 1
p_x_c_2_mamalia
```

```python
p_x_c_2_insekta = 0 * 1 * 1 * 0.208333 * 0.958333 * 1 * 0.916667
p_x_c_2_insekta
```

Menghitung Nilai P(C|X) sesuai rumus

```python
def prob_naive_bayes(p_x_c,p_c):
    rumus = p_x_c*p_c
    return rumus
```

## Menghitung Rumus Probabilitas dengan Metode Naive Bayes

Probabilitas Naive Bayes Pada Data Badak

```python
prob_naive_bayes(p_x_c_1_mamalia,0.5)
```

```python
prob_naive_bayes(p_x_c_1_insekta,0.5)
```

Probabilitas Naive Bayes Pada Data Kaki Seribu (Luwing)

```python
prob_naive_bayes(p_x_c_2_mamalia,0.5)
```

```python
prob_naive_bayes(p_x_c_2_insekta,0.5)
```

# 4. Bandingkan hasil probabilitas antara kelas mamalia dan insekta

```python
prob_naive_bayes(p_x_c_1_mamalia,0.5) > prob_naive_bayes(p_x_c_1_insekta,0.5)
```
