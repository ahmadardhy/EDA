Lakukan deteksi outlier untuk melihat potensi adanya fraud berdasarkan atribut Total amount Withdrawn (in Rupiah) pada dataset ATM Transaction menggunakan metode InterQuartileRange dan visualisasikan menggunakan BoxPlot serta sebutkan nama No Of Withdrawals dan nama ATM-nya yang terdeteksi fraud.

```python
Q1 = np.percentile(data4['Total amount Withdrawn (Rupiah)'], 25, interpolation = 'midpoint')
  
# Third quartile (Q3)
Q3 = np.percentile(data4['Total amount Withdrawn (Rupiah)'], 75, interpolation = 'midpoint')
  
# Interquaritle range (IQR)
IQR = Q3 - Q1
  
print(IQR)
```

```python
min_IQR = Q1 - 1.5 * IQR
max_IQR = Q3 + 1.5 * IQR

print('IQR minimum = ', min_IQR)
print('IQR maksimum = ', max_IQR)
```

```python
nilai_min = np.min(data4['Total amount Withdrawn (Rupiah)'])
nilai_max = np.max(data4['Total amount Withdrawn (Rupiah)'])

print('Nilai minimum data = ', nilai_min)
print('Nilai maksimum data = ', nilai_max)
```

```python
if (nilai_min < min_IQR):
    print('Ditemukan low outlier!')
else:
    print('Tidak ditemukan low outlier!')  

if (nilai_max > max_IQR):
    print('Ditemukan high outlier!')
else:
    print('Tidak ditemukan high outlier!')
```

```python
low_out = []
high_out = []

for i in data4['Total amount Withdrawn (Rupiah)']:
    if (i < min_IQR):
        low_out.append(i)
    if (i > max_IQR):
        high_out.append(i)

print('Low outlier : ', low_out)
print('High outlier : ', high_out)
```
