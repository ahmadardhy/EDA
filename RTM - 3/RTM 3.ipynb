### Membuat Fungsi Scratch untuk menghitung Persamaan Distribusi Uniform

```python
def dis_uniform(batas,a,b):
    atas = batas - a
    bawah = b - a
    rumus = atas/bawah
    return rumus
```

```python
dis_uniform(5,a,b)
```

### Membuat Fungsi Scratch untuk menghitung Persamaan Distribusi Normal

```python
import math
```

```python
def dis_normal(data, miu, gamma):
    phi = 3.141592
    e = 2.7183
    atas = 1*e**(-0.5*((data-miu)/gamma)**2)
    bawah = gamma*np.sqrt(2*phi)
    rumus = atas/bawah
    return rumus
```

```python
data2['f(x)'] = dis_normal(data2,0.079,2.317)
data2
```

```python
data_3 = pd.read_csv('./RTM-3 Dataset/skewness.csv')
data_3.head(5)
```

```python
data_3.describe()
```

```python
data_3['f(x)'] = dis_normal(data_3,-0.030769,2.734639)
data_3
```

```python
def skewness(data):
    mean = data.mean()
    median = data.median()
    atas = 3*(mean-median)
    bawah = data.std()
    rumus = atas/bawah
    return rumus
```

```python
skewness(data_3)
```

Visualisasi Diagram Distribusi Skewness

```python
import matplotlib.pyplot as plt

plt.plot(data_3['x'],data_3['f(x)'])
```

```python
data4['f(x)'] = dis_normal(data4,0.760465,12.474415)
data4.head(5)
```

Karena dibutuhkan sebuah hitungan yang menghitung `x-xbar`, maka di sini saya buat sebuah columns baru di sebelah kanan.

```python
data4['x-xbar'] = data4['x']-0.760465
data4
```

```python
sum(data4['x-xbar'])
```

```python
def kurtosis(data, n):
    total = sum(data4['x-xbar'])
    s = data.std()
    n = len(data)
    atas = 1/n*(total**4)
    bawah = s**4
    rumus = atas/bawah
    return rumus
```

```python
kurtosis(data4['x'],43)
```

Visualisasi data distribusi antara x dan f(x)

```python
import matplotlib.pyplot as plt

plt.plot(data4['x'],data4['f(x)'])
```
