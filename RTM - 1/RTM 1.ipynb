## No 1. Data Dosen

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_dosen = pd.read_csv('data dosen.csv')
data_dosen
```

```python
con1 = data_dosen['Tanggal Lahir']
data_dosen['Tanggal Lahir']=pd.to_datetime(data_dosen['Tanggal Lahir'])
```

```python
data_dosen
```

```python
from datetime import date

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
data_dosen['Umur'] = data_dosen['Tanggal Lahir'].apply(lambda x: calculate_age(x))
```

```python
data_dosen
```

```python
gol = data_dosen.groupby('Umur').size()
gol
```

### Plot Umur Dosen UPN Veteran X Angkatan 2022

```python
plt.figure(figsize=(10,4))
sns.set_style('darkgrid')
sns.countplot(x = 'Umur', data = data_dosen, color = "darkgreen")

plt.title('Umur Dosen UPN Veteran X Angkatan 2022')
plt.ylabel('Jumlah')

plt.savefig('Umur Dosen UPN Veteran X Angkatan 2022.png', dpi=300, bbox_inches='tight')
plt.show()
```

## No. 2 (Data Diabetes)

```python
data_diabetes = pd.read_csv('data diabetes.csv')
data_diabetes
```

```python
data_diabetes.describe()
```

Melakukan penghitungan BMI (Body Mess Index) untuk menentukan apakah berat badan dikategorikan kurang (underweight), normal, berat badan berlebih (overweight) atau berat badan sangat berlebih (obesitas)

```python
data_diabetes['Hasil BMI'] = ((data_diabetes['Berat Badan (kg)'])/((data_diabetes['Tinggi Badan (cm)'])/100)**2)
```

```python
data_diabetes
```

```python
batas_bin = [0, 17, 18.4, 25, 27,100]
kategori = ['Kurus - Tingkat Berat', 'Kurus - Tingkat Ringan', 'Normal','Gemuk - Tingkat Ringan','Gemuk - Tingkat Berat']
data_diabetes['Golongan BMI'] = pd.cut(data_diabetes['Hasil BMI'], bins=batas_bin, labels=kategori)
```

```python
data_diabetes
```

```python
count = data_diabetes.groupby('Golongan BMI').size()
count
```

```python
plt.figure(figsize=(12,5))
sns.set_style('whitegrid')
sns.countplot(x = 'Golongan BMI', data = data_diabetes, palette= "inferno")

plt.xlabel('Golongan BMI')
plt.ylabel('Jumlah')
plt.title('Golongan BMI Pasien Diabetes')

plt.savefig('Golongan BMI Pasien Diabetes.png', dpi=300, bbox_inches='tight')          
plt.show()
```

```python
import matplotlib.pyplot as plt
data_diabetes1 = data_diabetes.groupby('Golongan BMI').sum().reset_index()
labels = ['Kurus - Tingkat Berat', 'Kurus - Tingkat Ringan', 'Normal','Gemuk - Tingkat Ringan','Gemuk - Tingkat Berat']
plt.figure(figsize=(5,5))
plt.pie(count, labels = labels ,autopct='%1.2f%%')
plt.title('Persentase penderita diabetes berdasarkan golongan BMI')
plt.savefig('Persentase penderita diabetes berdasarkan golongan BMI.png', dpi=300, bbox_inches='tight')
plt.show()
```
